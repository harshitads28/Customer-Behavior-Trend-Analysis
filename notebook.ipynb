```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, f1_score
```

```python
models = {
    "Logistic Regression": log_model,
    "KNN": gridkn,
    "SVM": gridsvc,
    "Decision Tree": grid_tree,
    "Random Forest": grid_rfc,
}
```

```python
results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Single split metrics
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average="weighted")
    
    # Cross-validation
    cv_acc = cross_val_score(model, X, y, cv=5, scoring='accuracy').mean()
    cv_f1 = cross_val_score(model, X, y, cv=5, scoring='f1_weighted').mean()
    
    results.append([name, acc, f1, cv_acc, cv_f1])
```

```python
# Convert to DataFrame
df = pd.DataFrame(results, columns=["Model", "Accuracy", "F1", "CV Accuracy", "CV F1"]).set_index("Model")
```

```python
# Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(df, annot=True, cmap="YlGnBu", fmt=".3f")
plt.title("Model Performance Heatmap")
plt.show()
```

```python
data = {
    "Model": ["Logistic Regression", "KNN", "SVM", "Decision Tree", "Random Forest"],
    "Accuracy": [0.855, 0.860, 0.870, 0.845, 0.835],
    "F1": [0.802, 0.805, 0.810, 0.822, 0.799],
    "CV Accuracy": [0.881, 0.879, 0.883, 0.866, 0.871],
    "CV F1": [0.834, 0.846, 0.828, 0.836, 0.838]
}

df1 = pd.DataFrame(data).set_index("Model")

# Define weights
weights = {
    "Accuracy": 0.10,
    "F1": 0.20,
    "CV Accuracy": 0.30,
    "CV F1": 0.40
}

# Calculate weighted score
df1["Score"] = sum(df1[col] * weight for col, weight in weights.items())

# Rank models
df1["Rank"] = df1["Score"].rank(ascending=False)

# Display sorted results
df_sorted = df1.sort_values(by="Score", ascending=False)
print(df_sorted)
```
